```python
from langchain_huggingface.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
torch.set_default_device("cuda")
model = AutoModelForCausalLM.from_pretrained("microsoft/Phi-4-mini-instruct", dtype="auto")
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-4-mini-instruct")
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=500)
hf = HuggingFacePipeline(pipeline=pipe)
```

```python
hf.invoke("What is Capital of India?")
```

```python
hf.invoke("What is Capital of India?")
```

**What do you mean by huggingface pipeline here?**

A Hugging Face Pipeline is a high-level abstraction that allows you to easily use pre-trained models for various tasks without having to deal with the underlying model architecture or tokenization details.

```python
# chat_model via  HuggingFacePipeline
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline



llama_llm = HuggingFacePipeline.from_model_id(
    model_id="microsoft/Phi-4-mini-instruct",
    task="text-generation",
    device_map= "auto"
)

llama_llm_model = ChatHuggingFace(llm=llama_llm)
llama_llm_model.invoke("what is capital of US")
```

```python
res = llama_llm_model.invoke("what is capital of US")
```

```python
# To run a quantized version of your model, you can specify a bitsandbytes quantization config
from transformers import BitsAndBytesConfig
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline


quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype="float16",
    bnb_4bit_use_double_quant=True,
)

llm = HuggingFacePipeline.from_model_id(
    model_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    task="text-generation",
    pipeline_kwargs=dict(
        max_new_tokens=512
    ),
    model_kwargs={"quantization_config": quantization_config},
)
```
